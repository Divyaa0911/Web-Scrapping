```python
import requests #to fetch requests from given url 
import pandas as pd # it is easy to scarpe the web pages
from bs4 import BeautifulSoup # to parse and fetch the data which we want
```

```python
#web url
url='https://risk.lexisnexis.com/insights-resources/article/bank-rankings-top-banks-in-the-world'
```

```python
#Request page to get respnse
page=requests.get(url)
page
```

```python
# To display and fetch the content 
soup=BeautifulSoup(page.text)
soup.prettify
```

```python
# soup.find will find the table tag and id related to required table and store it in table variable
table=soup.find("table")
table
```

```python
rows=table.find_all('tr')
rows
```

```python
#we take an empty list and then apply for loop in rows which contains all the tr tags and after that we append the list we took as an empty list
column_info=[]
for i in rows:
    cols=i.find_all('td')
    #print('cols')
    bank_info=[c.text for c in cols]
    #print(bank_info)
    column_info.append(bank_info)
print(column_info)
```

```python
#  we take am empty list of all the columns and then apply for loop on that after that append that columns with index numbers
Current_Rank =[]
Previous_Rank =[]
Bank =[]
Assets_US=[]
local_currency=[]
Capital_US=[]
Balance_Sheet_Date=[]

a=[]
for i in column_info:
    a.append(i)

b=a[1:]

for i in b:
    Current_Rank.append(i[0])
    Previous_Rank.append(i[1])
    Bank.append(i[2])
    Assets_US.append(i[3])
    local_currency.append(i[4])
    Capital_US.append(i[5])
    Balance_Sheet_Date.append(i[6])
    
    
```

```python
# now for every column we take an empty list, the data is now present in another list
# we make dictinary of the above columns
Data={'Current_Rank':Current_Rank,'Previous_Rank':Previous_Rank,'Bank':Bank,'Assets_US':Assets_US,'local_currency':local_currency,'Capital_US':Capital_US,
     'Balance_Sheet_Date':Balance_Sheet_Date}
```

```python
# to make a dataframe 
bank=pd.DataFrame(Data)
bank
```

```python
# converting above dataframe into csv file
bank.to_csv('Top_Banks.csv')
```
